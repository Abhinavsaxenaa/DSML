```python
import pandas as pd
import numpy as np
df = pd.read_csv('terrorismData.csv')
dfc = df.copy()
dfc = dfc[dfc['State']=='Jammu and Kashmir']
dfc['Wounded'] = dfc['Wounded'].replace(np.nan, 0)
dfc['Killed'] = dfc['Killed'].replace(np.nan, 0)
dfc['TotalAttack'] = dfc['Killed'] + dfc['Wounded']
dfc = dfc[dfc['Group'] != 'Unknown']
dfc
```
